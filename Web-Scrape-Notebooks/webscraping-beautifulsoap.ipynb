```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup
```

```python
registration = []
color = []
assembly = []
capacity = []
body = []
for url in df['Url'][101793:]:
    response = requests.get(url)
    content = response.content
    read = BeautifulSoup(content, 'html.parser')
    details = read.find('ul', id = 'scroll_car_detail')
    
    if details != None:
        details = details.text.split('\n')
        items = [x for x in details if x]
        registration.append(items[1])
        color.append(items[3])
        assembly.append(items[5])
        capacity.append(items[7])
        body.append(items[9])
    else:
        registration.append('N/A')
        color.append('N/A')
        assembly.append('N/A')
        capacity.append('N/A')
        body.append('N/A') 
```
