## Feature Engineering

#### We have so many rows which has all the features as same but different prices. Why is that??
#### This may be because of the Condition of the car which is not already mentioned in our dataset.
#### So we have to introduce new Feature as 'Condition' of a car - like 'Excellent', 'Very Good', 'Good', 'Fair', 'Poor' so that we can have with all similar features but different 'Condition' results in different price.

```python
cols = ['Make', 'Model', 'Model_Year', 'Fuel_Type', 'Transmission', 'Engine_Capacity(cc)',
       'Kms_Driven', 'Color', 'Assembly', 'Body_Type']
cols_2 = ['Make', 'Model', 'Model_Year', 'Fuel_Type', 'Transmission', 'Engine_Capacity(cc)',
       'Kms_Driven', 'Color', 'Assembly', 'Body_Type', 'Price']
cols_3 = ['Make', 'Model', 'Model_Year', 'Fuel_Type', 'Transmission', 'Engine_Capacity(cc)',
       'Kms_Driven', 'Color', 'Assembly', 'Body_Type', 'Price', 'Condition']
cols_4 = ['Make', 'Model', 'Model_Year', 'Fuel_Type', 'Transmission', 'Engine_Capacity(cc)',
       'Kms_Driven', 'Color', 'Assembly', 'Body_Type', 'Condition']
```

```python
df[(df['Make'] == 'FAW') & (df['Model'] == 'X-PV') & (df['Model_Year'] == 2016) & (df['Fuel_Type'] == 'Petrol')
  & (df['Transmission'] == 'Manual') & (df['Engine_Capacity(cc)'] == 1000) & (df['Kms_Driven'] == 68000)
  & (df['Color'] == 'Grey') & (df['Assembly'] == 'Local') & (df['Body_Type'] == 'Mini Van') ]
```

##### Dropping Unecessary Columns

## Excellent

```python
max_grp = df[df.duplicated(cols, keep=False)].sort_values(by='Price').groupby(cols).max()
max_grp.reset_index(inplace=True)

max_grp['Condition'] = 'Excellent'
max_grp
```

```python
excellent = max_grp
```

```python
df = pd.concat([df,max_grp])
```

```python
df.drop(df[df[cols_2].duplicated(keep='last')].index, inplace=True)
```

```python
df.shape
```

```python
df =  df[df['Condition'] != 'Excellent']
```

## Very Good

```python
max_grp = df[df.duplicated(cols, keep=False)].sort_values(by='Price').groupby(cols).max()
max_grp.reset_index(inplace=True)

max_grp['Condition'] = 'Very Good'

very_good = max_grp

df = pd.concat([df,max_grp])

df.drop(df[df[cols_2].duplicated(keep='last')].index, inplace=True)

df =  df[df['Condition'] != 'Very Good']

# df[df['Condition'] == 'Very Good']
```

## Good

```python
max_grp = df[df.duplicated(cols, keep=False)].sort_values(by='Price').groupby(cols).max()
max_grp.reset_index(inplace=True)

max_grp['Condition'] = 'Good'

good = max_grp

df = pd.concat([df,max_grp])

df.drop(df[df[cols_2].duplicated(keep='last')].index, inplace=True)

df =  df[df['Condition'] != 'Good']

# df[df['Condition'] == 'Good']
```

## Fair

```python
max_grp = df[df.duplicated(cols, keep=False)].sort_values(by='Price').groupby(cols).min()
max_grp.reset_index(inplace=True)

max_grp['Condition'] = 'Fair'

fair = max_grp

df = pd.concat([df,max_grp])

df.drop(df[df[cols_2].duplicated(keep='last')].index, inplace=True)

df =  df[df['Condition'] != 'Fair']
```

## Poor

```python
max_grp = df[df.duplicated(cols, keep=False)].sort_values(by='Price').groupby(cols).min()
max_grp.reset_index(inplace=True)

max_grp['Condition'] = 'Poor'

poor = max_grp

df = pd.concat([df,max_grp])

df.drop(df[df[cols_2].duplicated(keep='last')].index, inplace=True)

df =  df[df['Condition'] != 'Poor']
```

```python
df = pd.concat([excellent,very_good,good,fair,poor,df])
```
